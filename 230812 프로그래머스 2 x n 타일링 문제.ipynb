이 문제는 동적 계획법과 모듈러 연산을 활용하여 푸는 문제이다.

모듈러 연산은 일반적으로 숫자의 범위를 제한하거나 큰 수를 다룰 때 오버플로우를 방지하고자 할 때 주로 사용된다. 순환 패턴을 갖는 문제에서(예-시저 암호) 주기적인 패턴을 찾을 때도 유용하게 사용될 수 있다.

```python
def solution(n):
    dp = [0] * (n + 1)
    dp[1] = 1
    dp[2] = 2
    
    for i in range(3, n + 1):
        dp[i] = (dp[i-1] + dp[i-2]) % 1000000007
        
    return dp[n] 
```

멀리 뛰기 문제와 같은 방식으로 풀었다.

문제의 조건은 가로의 길이가 2이고 세로의 길이가 1인 직사각형 타일을 사용하여 세로의 길이가 2이고 가로의 길이가 n인 바닥을 가득 채우려고 할 때 경우의 수를 구하는 것이다.

따라서 가로의 길이가 1일 때는 직사각형을 채우는 방법이 1가지밖에 없다. 즉, dp[1] = 1이다. 가로의 길이가 2일 때는 직사각형을 채우는 방법이 2가지가 존재한다. 가로로 2개의 타일을 배치하거나, 세로로 2개의 타일을 배치하는 경우이다. 따라서 dp[2] = 2이다.

이후의 dp[i]의 값은 이전의 두 값인 dp[i-1] + dp[i-2]가 되는데, 바로 앞의 두 가지 상황에서 파생되는 경우의 수를 더한 것이다. 

이때 경우의 수가 너무 커지면 오버플로우가 발생할 수 있으므로 문제에서 1000000007로 나누라고 하였다. 나는 이때 리턴할 때만 나눠주면 될 줄 알았는데 먼저 나눠주든 뒤에 나눠주든 상관은 없으나 리턴할 때 나눠줄 경우에는 런타임 에러가 발생하였다. 

어차피 나머지를 반환하는 문제이므로 반복문에서 나눠줘도 상관 없고, 1000000007보다 작기 때문에 그 수 그대로 저장된다.
